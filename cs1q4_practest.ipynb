```python

def averaget(s1, s2):
    return (s1 + s2) / 2

def averageth(s1, s2, s3):
    return (s1 + s2 + s3) / 3

def averagefo(s1, s2 ,s3 ,s4):
    return (s1 + s2 + s3 + s4) / 4

while True:
  str(input("Enter your name:"))
  amt = int(input("Enter the number of scores to be inputted (1-4): "))
  if amt == 1:
      s1 = int(input("Enter score: "))
      grade = s1
      if grade >= 75:
          print("You passed")
      else:
          print("You failed")
  elif amt == 2:
      s1 = int(input("Enter 1st score: "))
      s2 = int(input("Enter 2nd score: "))
      print("This is the average score:", averaget(s1, s2),)
      grade = averaget(s1, s2)
      if grade >= 75:
          print("You passed")
      else:
          print("You failed")
  elif amt == 3:
      s1 = int(input("Enter 1st score: "))
      s2 = int(input("Enter 2nd score: "))
      s3 = int(input("Enter 3rd score: "))
      print("This is the average score:", averageth(s1, s2, s3))
      grade = averaget(s1, s2)
      if grade >= 75:
          print("You passed")
      else:
          print("You failed")
  elif amt == 4:
      s1 = int(input("Enter 1st score: "))
      s2 = int(input("Enter 2nd score: "))
      s3 = int(input("Enter 3rd score: "))
      s4 = int(input("Enter 4th score: "))
      print("This is the average score:", averagefo(s1, s2, s3, s4))
      grade = averaget(s1, s2)
      if grade >= 75:
          print("You passed")
      else:
          print("You failed")
  else:
      print("Maximum is 4 score, or you put an invalid number")

  choice = input("compute again? (y/n):" )
  if choice == "y":
    pass
  else:
    break
```

Asks the user for their name
Asks the user for number of quiz scores to be inputted
Asks the user quiz scores repeatedly until maximum is reached
Computes the total and average score
Determines if the student passed or failed
Uses a function to organize the program
